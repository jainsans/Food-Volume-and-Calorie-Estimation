```python
import cv2
import numpy as np
from food_volume_estimation.food_segmentation.mrcnn.config import Config
from food_volume_estimation.food_segmentation.mrcnn.model import MaskRCNN

# Define configuration for the model
class InferenceConfig(Config):
    # Give the configuration a recognizable name
    NAME = "segmentation_of_food"

    # Set batch size to 1 since we're running inference on a single image
    GPU_COUNT = 1
    IMAGES_PER_GPU = 1
    # Number of classes (including background)
    NUM_CLASSES = 2  # 1 Background + 1 Object

# Create config object
config = InferenceConfig()

# Create Mask R-CNN model in inference mode
model = MaskRCNN(mode="inference", config=config, model_dir="./logs")

# Load pre-trained weights
model.load_weights(r"C:\Users\hp\Desktop\food_vol\mask_rcnn_food_segmentation.h5", by_name=True)
```

```python
import os
import cv2

# Folder containing the images
folder_path = r"C:\Users\hp\Desktop\neew\img"
output_path = r"C:\Users\hp\Desktop\neew\seg"

class_names = ['bg','food']

# Loop through all the files in the folder
for filename in os.listdir(folder_path):
    # Read the image using cv2.imread
    img = cv2.imread(os.path.join(folder_path, filename),cv2.IMREAD_COLOR)
    image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = model.detect([image], verbose=0)
    r = results[0]
    name = filename
    dot = name.find(".")
    name = filename[:dot]
    for i in range(r["rois"].shape[0]):
        masked_image = cv2.bitwise_and(image, image, mask=r["masks"][:,:,i].astype(np.uint8))
        fname = name + f"{i}.jpg"
        b,g,R = cv2.split(masked_image)
        rgb_image = cv2.merge([R,g,b])
        cv2.imwrite(os.path.join(output_path, fname),rgb_image)
#         print(fname,i)
#     print(filename)
```
